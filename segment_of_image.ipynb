```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
image = cv2.imread('second_image.jpg')
plt.imshow(image)
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# blur
blur = cv2.GaussianBlur(gray, (0,0), sigmaX=33, sigmaY=33)

# divide
divide = cv2.divide(gray, blur, scale=255)

# otsu threshold
thresh = cv2.threshold(divide, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]

# apply morphology
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(morph)
plt.show()
```

```python
(h,w)=morph.shape[:2]
```

```python
first,second,third=(w//4),(w//2),(3*w//4)
```

```python
first_image=morph[0:h,0:first]
```

```python
plt.imshow(first_image)
plt.show()
```

```python
second_image=morph[0:h,first:second]
```

```python
plt.imshow(second_image)
plt.show()
```

```python
third_image=morph[0:h,second:third]
plt.imshow(third_image)
plt.show()
```

```python
forth_image=morph[0:h,third:w]
plt.imshow(forth_image)
plt.show()
```
